```haskell
cuadrados :: [Int]
cuadrados = [x^2 | x <- [1..10]]
```

```haskell
cuadrados 
```

```haskell
-- 2. Divisibles por 3 entre 1 y 30
```

```haskell
divisiblesPor3 :: [Int]
divisiblesPor3 = [x | x <- [1..30], x `mod` 3 == 0]
```

```haskell
divisiblesPor3
```

```haskell
-- 3. Pares (x, y) donde x < y en [1..5]
```

```haskell
paresOrdenados :: [(Int, Int)]
paresOrdenados = [(x, y) | x <- [1..5], y <- [1..5], x < y]
```

```haskell
paresOrdenados
```

```haskell
-- Punto 2: Divisores
```

```haskell
divisores :: Int -> [Int]
divisores n = [d | d <- [1..n], n `mod` d == 0]
```

```haskell
divisores 8
```

```haskell
-- Punto 3: Perfecto
```

```haskell
divisoresPropios :: Int -> [Int]
divisoresPropios n = [d | d <- [1..n-1], n `mod` d == 0]
esPerfecto :: Int -> Bool
esPerfecto n = sum (divisoresPropios n) == n
```

```haskell
esPerfecto 6
```
