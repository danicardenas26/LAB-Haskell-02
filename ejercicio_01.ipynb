```haskell
-- PUNTO 1
-- 1. Longitud de lista recursivamente
```

```haskell
miLength :: [a] -> Int
miLength []     = 0
miLength (_:xs) = 1 + miLength xs
```

```haskell
miLength [1, 10, 100, 8, 9]
```

```haskell
-- 2. Ultimo elemento de una lista
```

```haskell
miLast :: [a] -> a
miLast [x]    = x
miLast (_:xs) = miLast xs
```

```haskell
miLast [1, 10, 100, 8, 9]
```

```haskell
-- 3. Duplicar elementos de una lista
```

```haskell
duplicar :: [Int] -> [Int]
duplicar []     = []
duplicar (x:xs) = (x * 2) : duplicar xs
```

```haskell
duplicar [1, 10, 100, 8, 9]
```

```haskell
-- 4. Eliminar pares de lista
```

```haskell
eliminarPares :: [Int] -> [Int]
eliminarPares [] = []
eliminarPares (x:xs)
    | odd x     = x : eliminarPares xs
    | otherwise = eliminarPares xs
```

```haskell
eliminarPares [1, 10, 100, 8, 9]
```

```haskell
-- Punto 2 (funciones de orden superior como map y filter)
-- 1. longitud de lista
```

```haskell
miLength' :: [a] -> Int
miLength' xs = sum [1 | _ <- xs]
```

```haskell
miLength' [1, 10, 100, 8, 9]
```

```haskell
-- 2. Ultimo elemento de lista
```

```haskell
miLast' :: [a] -> a
miLast' = foldl1 (\_ x -> x)
```

```haskell
miLast' [1, 10, 100, 8, 9]
```

```haskell
-- 3. Duplicar
```

```haskell
duplicar' :: [Int] -> [Int]
duplicar' = map (* 2)
```

```haskell
duplicar' [1, 10, 100, 8, 9]
```

```haskell
-- 4. Eliminar pares
```

```haskell
eliminarPares' :: [Int] -> [Int]
eliminarPares' = filter odd
```

```haskell
eliminarPares' [1, 10, 100, 8, 9]
```
