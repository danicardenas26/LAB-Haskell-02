```python
print("--- 1. PERMISO DE CONDUCCIÓN ---")

edad = int(input("Ingrese su edad: "))
tiene_licencia = input("¿Tiene licencia? (si/no): ").lower()
esta_sancionado = input("¿Tiene alguna sanción activa? (si/no): ").lower()

if edad >= 18 and tiene_licencia == 'si' and esta_sancionado == 'no':
    print("Resultado: ¡Puedes conducir!")
else:
    print("Resultado: NO puedes conducir. No cumples con los requisitos.")
```

```python
print(" 2. CICLO FOR Y ACUMULADORES ")

# 1 y 3. Suma de los primeros n números naturales usando un acumulador
n = 5
acumulador_suma = 0

print("Números pares encontrados:")
for i in range(1, n + 1):
    acumulador_suma += i  #

    # 2. Imprime solo los números pares usando if
    if i % 2 == 0:
        print(f"- El {i} es par")

print(f"Resultado final del acumulador (Suma de 1 a {n}): {acumulador_suma}")

# 4. Temperaturas mayores al promedio
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]
promedio = sum(temperaturas) / len(temperaturas)

# Filtramos usando un ciclo for simple (antes era por comprensión)
temp_altas = []
for temp in temperaturas:
    if temp > promedio:
        temp_altas.append(temp)

print(f"\nEl promedio de temperatura es: {promedio:.2f}°C")
print(f"Temperaturas mayores al promedio: {temp_altas}")
```

```python
print(" 3. CICLO WHILE ")

# 1. Modifica el ciclo para contar hacia atrás (ejemplo desde 5 hasta 1)
contador = 5
print("Cuenta regresiva:")
while contador > 0:
    print("Contador:", contador)
    contador -= 1

print("\n--- Bucle controlado por usuario ---")
# 2. Programa que solicita números hasta que ingrese 0
numero = -1

while numero != 0:
    numero = int(input("Ingresa un número (0 para detener el programa): "))
    if numero != 0:
        print(f"Ingresaste el {numero}.")
print("Programa terminado")
```

```python
print(" 4. PARES E IMPARES EN DICCIONARIO ")

suma_pares = 0
suma_impares = 0  # 1. Variable para acumular impares

for i in range(1, 11):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

# 2. Representa el estado final como un diccionario
estado_final = {
    "suma_pares": suma_pares,
    "suma_impares": suma_impares
}

print("Estado final del bucle:", estado_final)
```

```python
print(" 5. INVENTARIO CON VALIDACIONES ")

productos = []

while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")
    if nombre.lower() == "fin":
        break

    precio = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))

    # 1. Usa continue para saltar productos con precio negativo o cantidad menor a 0
    if precio < 0 or cantidad < 0:
        print(" -> [Error] El precio y la cantidad deben ser positivos. Producto ignorado.\n")
        continue

    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    print(" -> Producto agregado exitosamente.\n")

# Verificamos que la lista no esté vacía antes de hacer cálculos
if productos:
    total = 0
    for p in productos:
        total += p["precio"] * p["cantidad"]
    print(f"\nTotal del inventario: ${total:.2f}")

    # 2. Muestra el producto más caro usando un ciclo for simple
    producto_mas_caro = None
    max_precio = -1
    for p in productos:
        if p["precio"] > max_precio:
            max_precio = p["precio"]
            producto_mas_caro = p
    print(f"El producto más caro es: {producto_mas_caro['nombre']} (Precio: ${producto_mas_caro['precio']})")

    # 3. Representa el estado final del inventario como lista de diccionarios
    print("\nEstado final del inventario:")
    for p in productos:
        print(f" - {p}")
else:
    print("\nNo se registraron productos en el inventario.")
```
