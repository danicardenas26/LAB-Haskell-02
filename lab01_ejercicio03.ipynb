```python
import requests

print("--- 1. API DE USUARIOS (JSONPlaceholder) ---")

# Llamado a la API pública
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()

print("Usuarios cuyo correo electrónico termina en '.biz':\n")

# Procesamiento y filtrado imperativo
for u in usuarios:
    correo = u["email"]
    if correo.endswith(".biz"):
        print(f"- Nombre: {u['name']} | Correo: {correo}")
```

```python
import requests
import pandas as pd

print("--- 2. API DE POKÉMON (PokeAPI) ---")

nombre_pokemon = input("Ingrese el nombre de un Pokémon (ej. pikachu, charizard): ").lower().strip()

url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

if respuesta.status_code == 200:
    datos = respuesta.json()

    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]],
        "experiencia_base": datos.get("base_experience", "No registrada")
    }
    print("\n--- Información del Pokémon ---")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")

    print("-" * 30)

    # 1. Condicionales adicionales para Tipos
    if len(pokemon["tipos"]) > 1:
        print(f"Tipo: Pokémon de tipo mixto ({', '.join(pokemon['tipos'])})")
    else:
        print(f"Tipo: {pokemon['tipos'][0]}")

    # 2. Condicionales adicionales para Peso
    if pokemon["peso"] > 100:
        print("Categoría de peso: Pokémon de gran tamaño")
    else:
        print("Categoría de peso: Pokémon ligero")

    # 3. Uso adicional interesante mostrado en pantalla
    print(f"Experiencia base al derrotarlo: {pokemon['experiencia_base']} EXP")

    print("-" * 30)

    # 4. Usar 'for' para listar las habilidades una por una
    print("Habilidades:")
    for habilidad in pokemon["habilidades"]:
        print(f" {habilidad.capitalize()}")

    # Verificar si hay movimientos y mostrarlos (sin DataFrame)
    if 'moves' in datos and len(datos['moves']) > 0:
        print("\n Primeros Movimientos del Pokémon ")
        print("Aquí están algunos de los movimientos que puede aprender:")
        for i, move_data in enumerate(datos['moves'][:5]):
            move_name = move_data['move']['name'].replace('-', ' ').title()
            print(f" - {move_name}")
        if len(datos['moves']) > 5:
            print(f"...y {len(datos['moves']) - 5} movimientos más.")
    else:
        print("\n[!] No se encontraron movimientos para este Pokémon.")

else:
    print("\nPokémon no encontrado. Verifique el nombre ingresado (recuerde usar nombres en inglés, ej: squirtle).")
```
